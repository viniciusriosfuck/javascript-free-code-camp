```
%%javascript
// tabs is an array of titles of each site open within the window
var Window = function(tabs) {
  this.tabs = tabs; // We keep a record of the array inside the object
};

// When you join two windows into one window
Window.prototype.join = function (otherWindow) {
  this.tabs = this.tabs.concat(otherWindow.tabs);
  return this;
};

// When you open a new tab at the end
Window.prototype.tabOpen = function (tab) {
  this.tabs.push('new tab'); // Let's open a new tab for now
  return this;
};

// When you close a tab
Window.prototype.tabClose = function (index) {

  // Only change code below this line

  // when remove index, there is another count
  // so let's call splice referencing the original this.tabs with spread
  // and change only when joining the indexes

  // console.log(this.tabs, index)
  var tabsBeforeIndex = [...this.tabs].splice(0, index); // Get the tabs before the tab
  var tabsAfterIndex = [...this.tabs].splice(index+1); // Get the tabs after the tab
  // console.log(tabsBeforeIndex, tabsAfterIndex)
  this.tabs = tabsBeforeIndex.concat(tabsAfterIndex); // Join them together
  // console.log(this.tabs)

  // Only change code above this line

  return this;
 };

// Let's create three browser windows
var workWindow = new Window(['GMail', 'Inbox', 'Work mail', 'Docs', 'freeCodeCamp']); // Your mailbox, drive, and other work sites
var socialWindow = new Window(['FB', 'Gitter', 'Reddit', 'Twitter', 'Medium']); // Social sites
var videoWindow = new Window(['Netflix', 'YouTube', 'Vimeo', 'Vine']); // Entertainment sites

// Now perform the tab opening, closing, and other operations
var finalTabs = socialWindow
  .tabOpen() // Open a new tab for cat memes
  .join(videoWindow.tabClose(2)) // Close third tab in video window, and join
  .join(workWindow.tabClose(1).tabOpen());
console.log(finalTabs.tabs);
```

```
%%javascript
// https://www.freecodecamp.org/learn/javascript-algorithms-and-data-structures/basic-algorithm-scripting/confirm-the-ending
function confirmEnding(str, target) {
  let check = true;
  for (let i=0; i<target.length; i++) {
    if (str[str.length-target.length+i] != target[i]) {
      check = false
      console.log(str[str.length-target.length+i], target[i])
    }
  }
  return check;
}

confirmEnding("Bastian", "n");
```

```
%%javascript
function repeatStringNumTimes(str, num) {
  if (num <= 0) {
    return '';
  } else {
    let arr = [];
    for (let i=0; i<num; i++) {
      arr.push(str)
    }
    return arr.join('');

  }  
}

repeatStringNumTimes("abc", 3);
console.log(repeatStringNumTimes("abc", 3));
```

```
%%javascript
function truncateString(str, num) {
  if (str.slice(0,num).length >= str.length) {
    return str
  } else {
    return str.slice(0,num)+'...'
  }
}

truncateString("A-tisket a-tasket A green and yellow basket", 8);
console.log(truncateString("A-tisket a-tasket A green and yellow basket", 8));
```

```
%%javascript
function getIndexToIns(arr, num) {

  // empty array
  if (arr.length === 0) {
    return 0
  }

  // numeric sort
  // sort return in ASCII ascending order
  // Ex: 1 comes before 2, so arr = [2, 10] sorted is [10, 2]
  let arrSorted = [...arr].sort( (a, b) => a-b );
  // console.log(arrSorted)

  // num within arr range
  for (let index in arrSorted) {
    if (num <= arrSorted[index]) {
      // console.log(typeof index)  // string
      return index*1;  // convert str to number
    }
  }

  // num out of arr range  
  return arr.length;
}

// getIndexToIns([40, 60], 50);
console.log(getIndexToIns([5, 3, 20, 3], 5))
```

```
%%javascript

function chunkArrayInGroups(arr, size) {
  const rows = Math.ceil(arr.length / size)
  // console.log(rows)

  let arrSplitted = [];
  for (let i=0; i<rows; i++) {
    arrSplitted.push(arr.slice(i*size, (i+1)*size ))
  }
  // console.log(arrSplitted)

  return arrSplitted;
}

chunkArrayInGroups(["a", "b", "c", "d"], 2);

```

```
%%javascript

function diffArray(arr1, arr2) {
  const s1 = new Set(arr1);
  const s2 = new Set(arr2);
  const diff12 = new Set(
    [...s1].filter(x => !s2.has(x)));
  const diff21 = new Set(
    [...s2].filter(x => !s1.has(x)));
  const symDiff = [...diff12, ...diff21];
  return symDiff;
}

console.log(diffArray([1, 2, 3, 5], [1, 2, 3, 4, 5]));
```

## Pig Latin (vowels x consonants)

```
%%javascript

function translatePigLatin(str) {
  let str0 = str.slice();
  const regexPatternVowel = /^[aeiou]/;
  const regexPatternConsonant = /^([^aeiou]+)(.*)$/;
  if (str0.match(regexPatternVowel)) {
    str0 = str0 + 'way'
  } else {
    str0 = str0.replace(regexPatternConsonant, '$2$1ay')
  }


  return str0;
}

translatePigLatin("consonant");
```

## Matched case replace

```
%%javascript

function myReplace(str, before, after) {

  const regexPattern = /^[A-Z]/;

  if (before.match(regexPattern)) {
    after = after[0].toUpperCase() + after.slice(1,)
  } else {
    after = after.toLowerCase()
  }
  let str0 = str.replace(before,after);

  return str0;
}

myReplace("A quick brown fox jumped over the lazy dog", "jumped", "leaped");
```

## Fibonacci

```
%%javascript

function fib(num) {
  if ( (num == 1) | (num == 2) ) {
    return 1;
  } else {
    return fib(num-1) + fib(num-2);
  }
}

// console.log(fib(7))

function sumFibs(num) {
  let sum = 0;
  let fibI;
  let i=1;

  while (true) {
    fibI = fib(i);
    if (fibI > num) {
      break
    }; 
    if (fibI % 2 != 0) {
      sum += fibI;
    };
    i++;
  }  
  return sum;
}
// console.log(sumFibs(75025))

sumFibs(4);
```
